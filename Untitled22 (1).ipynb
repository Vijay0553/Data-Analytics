```python
import pandas as pd
from IPython.display import display

# Step 1: Load the Excel file
file_path = "NULL FILE.xlsx"
df = pd.read_excel(file_path)

# Step 2: Remove rows where 'Product_id' is missing
df = df[df['Product_id'].notna()]

# Step 2.5: Remove rows where both 'BrandName' and 'Category' are missing
df = df[~(df['BrandName'].isna() & df['Category'].isna())]

# Step 3: Fill missing 'OriginalPrice (in Rs)' with the average from same 'Category'
df['OriginalPrice (in Rs)'] = df['OriginalPrice (in Rs)'].astype(float)
df['OriginalPrice (in Rs)'] = df.groupby('Category')['OriginalPrice (in Rs)']\
                                .transform(lambda x: x.fillna(x.mean()))

# Step 4: Fill missing 'DiscountPrice (in Rs)' using 'DiscountOffer' and 'OriginalPrice (in Rs)'
# First, convert 'DiscountOffer' (like "40% OFF") into numeric percentage (like 0.40)
df['DiscountOffer'] = df['DiscountOffer'].str.extract(r'(\d+)', expand=False).astype(float) / 100

# Now fill missing DiscountPrice using formula: OriginalPrice * (1 - Discount%)
df['DiscountPrice (in Rs)'] = df.apply(
    lambda row: row['OriginalPrice (in Rs)'] * (1 - row['DiscountOffer']) 
    if pd.isna(row['DiscountPrice (in Rs)']) and pd.notna(row['DiscountOffer']) 
    else row['DiscountPrice (in Rs)'],
    axis=1
)

# Step 5: Display the cleaned DataFrame
print("Cleaned data with missing Discount Prices filled using DiscountOffer percentage:")
display(df)

# Step 6: Save the cleaned data to a new Excel file
df.to_excel("Cleaned_Priced_NULL_FILE.xlsx", index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Apply dark theme for the plot
plt.style.use('seaborn-v0_8-dark')

# Load the cleaned Excel file
df = pd.read_excel("Cleaned_Priced_NULL_FILE.xlsx")

# Step 1: Frequency Distribution - Orders per Brand
brand_counts = df['BrandName'].value_counts().sort_values(ascending=False)
frequency_distribution = brand_counts.reset_index()
frequency_distribution.columns = ['BrandName', 'OrderCount']

# Print frequency distribution in console
print(" Frequency Distribution of Orders by Brand:")
print(frequency_distribution)

# Optional: Save the table to a new Excel file
frequency_distribution.to_excel("Frequency_Distribution_BrandOrders.xlsx", index=False)

# Step 2: Plot the histogram-style bar chart
colors = []
for i in range(len(brand_counts)):
    if i == 0:
        colors.append('limegreen')  # Top brand
    elif i == len(brand_counts) - 1:
        colors.append('crimson')    # Bottom brand
    else:
        colors.append('dodgerblue') # Others

plt.figure(figsize=(14, 7))
bars = plt.bar(brand_counts.index, brand_counts.values, color=colors, edgecolor='black')

# Add value labels on top of bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height + 1, f'{int(height)}', 
             ha='center', va='bottom', fontsize=9, color='white')

# Annotate highest and lowest
plt.annotate('Highest Orders', 
             xy=(0, brand_counts.values[0]), 
             xytext=(0, brand_counts.values[0] + 10),
             arrowprops=dict(facecolor='limegreen', shrink=0.05),
             ha='center', fontsize=10, color='white')

plt.annotate('Lowest Orders',
             xy=(len(brand_counts)-1, brand_counts.values[-1]),
             xytext=(len(brand_counts)-1, brand_counts.values[-1] + 10),
             arrowprops=dict(facecolor='crimson', shrink=0.05),
             ha='center', fontsize=10, color='white')

# Plot labels and styling
plt.xlabel("Brand Name", fontsize=12, color='white')
plt.ylabel("Number of Orders", fontsize=12, color='white')
plt.title("Histogram of Orders by Brand (Dark Theme)", fontsize=14, color='white')
plt.xticks(rotation=45, ha='right', color='white')
plt.yticks(color='white')
plt.tight_layout()

# Show plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('seaborn-v0_8-dark')
df = pd.read_excel("Cleaned_Priced_NULL_FILE.xlsx")
grouped = df.groupby(['Category', 'BrandName']).size().reset_index(name='OrderCount')
top_brands = grouped.loc[grouped.groupby('Category')['OrderCount'].idxmax()].sort_values('Category')
low_brands = grouped.loc[grouped.groupby('Category')['OrderCount'].idxmin()].sort_values('Category')
categories = top_brands['Category'].tolist()
x = np.arange(len(categories))
plt.figure(figsize=(15, 7))
plt.plot(x, top_brands['OrderCount'], label='Most Ordered Brand', marker='o', color='limegreen', linewidth=2)
plt.plot(x, low_brands['OrderCount'], label='Least Ordered Brand', marker='o', color='crimson', linewidth=2)
for i, (brand, count) in enumerate(zip(top_brands['BrandName'], top_brands['OrderCount'])):
    plt.text(x[i], count + 1, f"{brand}", ha='center', va='bottom', fontsize=8, color='limegreen')
for i, (brand, count) in enumerate(zip(low_brands['BrandName'], low_brands['OrderCount'])):
    plt.text(x[i], count - 3, f"{brand}", ha='center', va='top', fontsize=8, color='crimson')
plt.xticks(x, categories, rotation=45, ha='right', color='white')  
plt.yticks(color='white')                                         
plt.xlabel("Product Categories", color='white', fontsize=12)       
plt.ylabel("Number of Orders", color='white', fontsize=12)        
plt.title("Most vs Least Ordered Brands in Each Category", color='white', fontsize=15)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
data = {
    'Category': ['Western', 'Topwear', 'Indian Wear', 'Sports Wear', 'Lingerie', 'Inner Wear'],
    'ActualCount': [2000, 1200, 500, 350, 100, 150],
    'PredictedCount': [1738, 586, 826, 250, 350, 150]
}

df = pd.DataFrame(data)
df['Error'] = np.abs(df['ActualCount'] - df['PredictedCount'])
x = np.arange(len(df))
plt.figure(figsize=(12, 6))
plt.plot(x, df['ActualCount'], marker='o', linestyle='-', label='Actual Orders', color='royalblue')
plt.plot(x, df['PredictedCount'], marker='o', linestyle='--', label='Predicted Orders', color='darkorange')
plt.fill_between(x, df['ActualCount'], df['PredictedCount'], color='gray', alpha=0.3, label='Prediction Error')
plt.xticks(x, df['Category'], rotation=45)
plt.xlabel("Product Category", fontsize=12)
plt.ylabel("Order Count", fontsize=12)
plt.title("Actual vs Predicted Order Counts by Category", fontsize=14)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
file_path = 'NULL FILE.xlsx'
df = pd.read_excel(file_path)
print("Original Data Sample:")
print(df.head())
df.columns = df.columns.str.strip().str.replace(' ', '_').str.replace(r'[^\w\s]', '', regex=True)
df['BrandName'].fillna('Unknown', inplace=True)
df['category_by_Gender'].fillna('Unspecified', inplace=True)
df['Calculated_Discount_Percentage'] = ((df['OriginalPrice_in_Rs'] - df['DiscountPrice_in_Rs']) / df['OriginalPrice_in_Rs']) * 100
df['Calculated_Discount_Percentage'] = df['Calculated_Discount_Percentage'].round(2)
summary = df.groupby(['category_by_Gender', 'Category']).agg({
    'Product_id': 'count',
    'DiscountPrice_in_Rs': 'mean',
    'Calculated_Discount_Percentage': 'mean'
}).reset_index()

summary.rename(columns={
    'Product_id': 'Product_Count',
    'DiscountPrice_in_Rs': 'Avg_Discounted_Price',
    'Calculated_Discount_Percentage': 'Avg_Discount_%'
}, inplace=True)
print("\nTransformed Summary:")
print(summary)
summary.to_excel('transformed_summary.xlsx', index=False)
```

```python
import pandas as pd
df = pd.read_excel("Cleaned_Priced_NULL_FILE.xlsx")
brand_counts = df['BrandName'].value_counts().sort_values(ascending=False)
frequency_distribution = brand_counts.reset_index()
frequency_distribution.columns = ['BrandName', 'OrderCount']
mean_orders = brand_counts.mean()
median_orders = brand_counts.median()
std_orders = brand_counts.std()
print(" Frequency Distribution of Orders by Brand:")
print(frequency_distribution)

print("\n Statistical Summary:")
print(f"Mean Orders per Brand: {mean_orders:.2f}")
print(f"Median Orders per Brand: {median_orders:.2f}")
print(f"Standard Deviation of Orders: {std_orders:.2f}")
frequency_distribution.to_excel("Brand_Orders_Distribution.xlsx", index=False)

stats_df = pd.DataFrame({
    'Statistic': ['Mean', 'Median', 'Standard Deviation'],
    'Value': [mean_orders, median_orders, std_orders]
})
stats_df.to_excel("Brand_Orders_Statistics.xlsx", index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_excel("Cleaned_Priced_NULL_FILE.xlsx")
df.columns = df.columns.str.strip().str.replace(' ', '_').str.replace(r'[^\w\s]', '', regex=True)
numeric_df = df.select_dtypes(include=['int64', 'float64'])
numeric_df_cleaned = numeric_df.dropna()
correlation_matrix = numeric_df_cleaned.corr()
plt.figure(figsize=(10, 8))
sns.set(style="whitegrid")
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of Numeric Features", fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
